```python
## importing libray for reading images from multiple folders: 
import cv2
import numpy
import glob
```

```python
folder_food = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Food/*')
folder_accessories = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Accessories/*')
folder_Clothes = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Clothes/*')
folder_Cosmetic = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Cosmetic/*')
folder_Electronic = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Electronic/*')
folder_Institution = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Institution/*')
folder_Leisure = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Leisure/*')
folder_Medical = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Medical/*')
folder_Necessities = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Necessities/*')
folder_Transportation = glob.glob('E:/SEM2/Dessertation/Face_detection_project/Dataset/Logo_dataset/Logo-2K+/Transportation/*')


Rez_size = 2
image_shape = 32 * Rez_size # rows/cols (should be square)
no_channel = 3
```

```python
logo_food = []
for folder in folder_food:
    for filename in tqdm(os.listdir(folder)):
            path = os.path.join(folder,filename)
            in_image = Image.open(path).resize((image_shape,
                  image_shape),Image.ANTIALIAS)
            logo_food.append(np.asarray(in_image))
logo_food = np.reshape(logo_food,(-1,image_shape,
            image_shape,no_channel))
logo_food = logo_food.astype(np.float32)
logo_food = logo_food / 127.5 - 1.
```
